Імпорт необхідних модулів

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

1. Вивести перші рядки таблиці за допомогою методу head

```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
table = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', decimal=',', thousands=".")
df = table[0]
df.head() # Метод df.head() у Python використовується для виводу перших п'яти рядків DataFrame
```

2.Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
num_rows, num_columns = df.shape

print(f"Кількість рядків: {num_rows}")
print(f"Кількість стовбців: {num_columns}")
```

3.Замініть у таблиці значення "—" на значення NaN

```python
new_table = df.replace("—", np.nan)
new_table
```

4.Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df.dtypes
```

5.Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

Пояснення коду написаного нижче :

df.columns[1:] - Отримує всі назви стовпців, починаючи з другого.  
df[df.columns[1:]]: -Вибирає всі стовпці, починаючи з другого
 з DataFrame df.apply(pd.to_numeric, errors='coerce') - 
Застосовує функцію pd.to_numeric до вибраних стовпців, перетворюючи їх у числовий формат.   
Параметр errors='coerce' означає, що в разі помилки перетворення значення будуть замінені на NaN.

```python
df[df.columns[1:]] = df[df.columns[1:]].apply(pd.to_numeric, errors='coerce')
df.dtypes
```

6.Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

Алгоритм обрахунку :   
   Формула кількість пустих кліток(NaN) в совпці ділимо на кількість всіх кліток і множимо на 100 отримуємо відсоток пустих кліток в кожному стовпці 

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100 
missing_percentage_rounded = missing_percentage.round(1)
print(missing_percentage_rounded)
```

7.Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df.drop(df.index[-1], inplace=True)
df.tail()  # Метод df.tail() у Python використовується для виводу останніх п'яти рядків DataFrame
```

8.Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df_mean=df.fillna(df.mean(numeric_only=True))
df_mean.round(2)
```

9.Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
YEAR = '2019'
COL_NAME = 'Регіон'
condition = df[YEAR] > df.mean(numeric_only=True)[YEAR]
print("Cписок регіонів, де народжуваності у 2019 році був вищим за середній по Україні :")
df[condition][COL_NAME]
```

10.У якому регіоні була найвища народжуваність у 2014 році?

```python
YEAR = '2014'
df.loc[df[YEAR].idxmax(), COL_NAME]
```

11.Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
YEAR= '2019'
plt.style.use('ggplot')
plt.figure(figsize=(10, 6)) 
plt.bar(df[COL_NAME], df[YEAR], color='g')
plt.xticks(rotation=60, ha='right')
plt.ylabel('Коефіцієнт народжуваності у 2019 році')
plt.legend(['2019 рік'])
plt.show()
```

```python
lviv = df[df['Регіон'] == 'Львівська']
plt.plot(lviv.columns[1:], lviv.iloc[0,1:], marker='o', markerfacecolor='red')
plt.title('Динаміка народжуваності у Львівській області')
plt.xlabel('Рік')
plt.ylabel('Кількість народжених (на 1000 осіб)')
plt.show()
```

```python
plt.pie(
    df_mean['2014'].loc[0:26],
    labels=df_mean['Регіон'].loc[0:26],
    shadow=True,
    autopct="%.2f%%",
    pctdistance=1.37,
    labeldistance=1.59,
)
plt.title('Динаміка народжуваності по областям за 2014 рік', pad=60)
plt.show()
```
