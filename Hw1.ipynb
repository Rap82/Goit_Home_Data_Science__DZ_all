# 8. Створіть матрицю розміром 3x3 , заповніть її випадковими цілими числами в діапазоні від 1 до 10 та знайдіть її обернену матрицю.

https://uk.wikipedia.org/wiki/%D0%9E%D0%B1%D0%B5%D1%80%D0%BD%D0%B5%D0%BD%D0%B0_%D0%BC%D0%B0%D1%82%D1%80%D0%B8%D1%86%D1%8F#:~:text=%D0%9E%D0%B1%D0%B5%D1%80%D0%BD%D0%B5%D0%BD%D0%B0%20%D0%BC%D0%B0%D1%82%D1%80%D0%B8%D1%86%D1%8F%20%E2%80%94,%D0%BC%D0%B0%D1%82%D1%80%D0%B8%D1%86%D1%8F%20%D1%94%20%D0%BD%D0%B5%D0%B2%D0%B8%D1%80%D0%BE%D0%B4%D0%B6%D0%B5%D0%BD%D0%BE%D1%8E.

```python
# Створимо матрицю розміром 3x3 з випадковими цілими числами від 1 до 10
matrix = np.random.randint(1, 11, (3, 3))

# Виведемо початкову матрицю
print("Початкова матриця:")
print(matrix)

# Знайдемо обернену матрицю
try:
    inverse_matrix = np.linalg.inv(matrix) # Функція повертає обернену матицію до вказаної якщо вона існує і накше повертає помилку *np.linalg.LinAlgError
    print("Обернена матриця:")
    print(inverse_matrix)
except np.linalg.LinAlgError:
    print("Матриця не має оберненої матриці, оскільки вона може бути сингулярною.")
```

# 16. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.

```python
# Створимо матрицю 3x3 з випадковими дійсними числами в діапазоні від 0 до 1
matrix = np.random.rand(3, 3)

# Виведемо матрицю
print("Матриця:")
print(matrix)

# Знайдемо суму елементів кожного рядка матриці та створимо вектор-стовпчик
sum_of_rows = np.sum(matrix, axis=1, keepdims=True) # Функція np.sum() в бібліотеці Numpy використовується для обчислення суми елементів масиву або матриці.
# Де:
    # a - вхідний масив або матриця
    # axis - вісь, за якою обчислюється сума. Якщо не вказано, сума обчислюється для всіх елементів.
    # keepdims - якщо встановлено в True, то вимірність результуючого масиву зберігається, навіть якщо розмірність за вказаною віссю дорівнює 1.

# Виведемо вектор-стовпчик із сумами
print("Вектор-стовпчик із сумами рядків:")
print(sum_of_rows)
```
